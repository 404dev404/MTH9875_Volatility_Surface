##The Kou double-exponential jump diffusion model


In the Kou model, the underlying follows a jump-diffusion process of the form

$$
x_t=\alpha\,t+\sigma\,W_t+\sum_{i=1}^{N_t}\,y_i
$$

where $\{N_t\}$ is the Poisson process counting the jumps of $x$ and $y_i$ are jump sizes.  The distribution of jump sizes is an asymmetric exponential with density

<a name="eq:KouDensity"></a>(1)
$$
\nu(x)=p\,\lambda_+\,e^{-\lambda_+\,x}\,{\bf 1}_{x>0}+(1-p)\,\lambda_-\,e^{+\lambda_-\,x}\,{\bf 1}_{x \leq 0}.
$$

###2. (8 points) 

Using the Lévy Khintchine formula, derive the characteristic function 
$
\phi_T(u)=\exp\{\psi(u)\,T\}
$
with

<a name="eq:KouPsi"></a>(2)
$$
\psi(u)=-\frac{\sigma^2}{2}\,u^2+i\,\omega\,u+i\,u\,\lambda\,\left\{\frac{p}{\lambda_+-i\,u}-\frac{1-p}{\lambda_-+i\,u}\right\}
$$

where $\lambda$ is the mean jump arrival rate and

$$
\omega=-\frac{\sigma^2}{2}-\lambda\,\left\{\frac{p}{\lambda_+-1}-\frac{1-p}{\lambda_-+1}\right\}.
$$

$\psi(u)$ is known as the *characteristic exponent* of the Lévy process.

2.

The Levy Khintchine formula:
$
\phi_T(u)=\exp\left\{i \,u\, \omega\, T -\frac{1}{2}u^2\, \sigma^2 T +T
\int \left[e^{i \,u \,\xi}-1\right]\mu(\xi)\,d\xi\right\}
$

For $\mu(\xi) =\lambda(p\,\lambda_+\,e^{-\lambda_+\,\xi}\,{\bf 1}_{\xi>0}+(1-p)\,\lambda_-\,e^{+\lambda_-\,\xi}\,{\bf 1}_{\xi \leq 0})
$

Plug in and integrate:
$
\int \left[e^{i \,u \,\xi}-1\right]\mu(\xi)\,d\xi = \lambda(\frac{p\lambda_+}{\lambda_+-iu}+\frac{(1-p)\lambda_-}{\lambda_-+iu}-1) = i\,u\,\lambda\left\{\frac{p}{\lambda_+-i\,u}-\frac{1-p}{\lambda_-+i\,u}\right\}
$

Therefore
$
\phi_T(u)=\exp\{T(-\frac{\sigma^2}{2}\,u^2+i\,\omega\,u+i\,u\,\lambda\,\left\{\frac{p}{\lambda_+-i\,u}-\frac{1-p}{\lambda_-+i\,u}\right\})\}
$

So when $\psi(-i) = 0$:
$$
\omega=-\frac{\sigma^2}{2}-\lambda\,\left\{\frac{p}{\lambda_+-1}-\frac{1-p}{\lambda_-+1}\right\}.
$$

###3. (6 points) 

Use the characteristic exponent [(2)](#eq:KouPsi) and Lewis's formula (6) from the notes to plot the one-year implied volatility smile assuming the following parameters:

$$
\sigma=0.2; \,\lambda=1.0;\,p=0.25;\,\lambda_-=8;\,\lambda_+=25;
$$

The log-strike range should be from $k=-1.0$ to $k=1.0$.  

```r
source("Heston2.R")
```

```r
#(3)
Kou_Char <- function(params) #characteristic function of Kou process
{
    sig <- params$sig
    lambda <- params$lambda
    p <- params$p
    lambdaNeg <- params$lambdaNeg
    lambdaPos <- params$lambdaPos
       
    res <- function(u, t)
    {
        omega <- -sig * sig / 2.0 - lambda * (p / (lambdaPos - 1) - (1-p) / (lambdaNeg + 1))      
        tmp = 1i * u * lambda * (p / (lambdaPos - 1i * u) - (1 - p) / (lambdaNeg + 1i * u))     
        phi = -0.5 * (sig ^ 2) * (u ^ 2) + 1i * omega * u + tmp   #given above in #2     
        res = exp(phi * t) #implied vol skew    
        return(res)
    }  
    return(res) #implied vol skew
}

Kou_Param <- list(sig = 0.2, lambda = 1., p = 0.25, lambdaNeg = 8., lambdaPos = 25.) #given paramters in problem

T <- 1 #1 year

vol <- function(k){sapply(k,function(x){bsvol(Kou_Char(Kou_Param),x,T)})} # Vectorize the function

curve(vol(x),from=-1,to=1,col='blue', xlab="Log strike (K)", ylab="Imp vol") #graph vol from k = -1 to 1
```

##Jump-Diffusion Monte Carlo

###4. (10 points) 

Write Monte Carlo code to price European options in the Kou model.  With the same parameters as in Problem 3, plot the one-year implied volatility smile from $k=-1.0$ to $k=1.0$.  Verify that you obtain the same implied volatilities.

```r
#(4)
Jump_Sum <- function(N_t){ #calculate sum of jumps in Kou model
    prob <- runif(N_t, min=0, max=1)
    exp_pos <- rexp(N_t, rate=lambdaPos)
    exp_neg <- -1. * rexp(N_t, rate=lambdaNeg)
    jumps <- (prob < p) *  exp_pos + (prob >= p) * exp_neg
    return(sum(jumps))
}

MC <- function(params){ #monte carlo simulation to return imp vol
        
    res <- function(N, T = 1.)
    {
        sig <<- params$sig
        lambda <<- params$lambda
        p <<- params$p
        lambdaNeg <<- params$lambdaNeg
        lambdaPos <<- params$lambdaPos
  
        alpha <- -lambda * ((p / lambdaPos - (1 - p) / lambdaNeg))  
        alpha <- 0
        W <- rnorm(N) * sqrt(T)
        Nts <- rpois(N, lambda * T)
        jumps <- sapply(Nts, function(N_t){Jump_Sum (N_t)})
        x <- alpha * T + sig * W + jumps
        x <- x - log(mean(exp(x))) #turns all number to positivie
        
        return(exp(x))
    }
        
    return(res)
}

MC_Call <-function(x, K){ #payoff
            
    V <- (x > K)*(x - K)  # Boundary condition
    
    return(mean(V))
}

N <- 10000 #10,000 paths

S_T <- MC(Kou_Param)(N)

MC_Calls <- function(k){sapply(k, function(x){MC_Call(S_T, exp(x))})} #vectorized the functions so we can graph it
Kou_Calls <- function(k){sapply(k, function(x){callOption(Kou_Char(Kou_Param), x, 1.0)})}
volMC <- function(k){sapply(k, function(x){BSImpliedVolCall(1, exp(x), T, 0, MC_Call(S_T, exp(x)))})}

vol <- function(k){sapply(k,function(x){bsvol(Kou_Char(Kou_Param),x,1)})}

curve(volMC(x),from=-1.0,to=1.0,col="blue", xlab="Log Strike)", ylab='Implied Volatility') #graphs
curve(vol(x),from=-1,to=1,col="red", add=TRUE)
legend('bottomleft', c("Close form","Simulation") , lty=1, col=c("red", "blue"))
```

##Term structure of ATM skew

###5.(4 points) 

Using formula (8) from the lecture notes:

\begin{equation}
\left.\frac{\partial \sigma_{BS} }{\partial k }\right|_{k=0}=-
e^{\frac{\sigma_{BS}^2 T}{8}} \sqrt{\frac{2}{ \pi}}\frac{1}{
\sqrt{T}}\,\int_0^\infty\,du \, \frac{u\,
\mathrm{Im}\left[\phi_T(u-i/2)\right]}
{u^2+\frac{1}{4}},
\end{equation}
plot the ATM volatility skew

$$
\left.\frac{\partial \sigma_{BS} }{\partial k }\right|_{k=0}
$$

from 0 to 10 years.  Superimpose the curve $a/\sqrt{T}$, setting $a$ so as to match the one-year ATM skew. 

```r
#(5)
ATM_Skew <- function(phi, t){ #sum of total jump
    atmVol <- bsvol(phi,0,t)   
    integrand <-  function(u){Im(u*phi(u - 1i/2, t)/(u^2 + 1/4))}    
    res <- -1. * exp(atmVol^2*t/8) * sqrt(2/pi) / sqrt(t) * integrate(integrand, lower = 0, 
        upper = Inf, rel.tol = 1e-10, subdivisions = 1000)$value  
    return(res)
}

skew <- function(T){sapply(T,function(x){ATM_Skew(Kou_Char(Kou_Param),x)})} #vectorize BS_Skew

alpha <- ATM_Skew(Kou_Char(Kou_Param), 1) #alpha = 1y atm skew

curve(skew(x),from=0.01,to=10,col="blue", xlab="Time", ylab="ATM Skew") #graph atm skew
curve(alpha / sqrt(x), from=0.01,to=10,col="red", add=TRUE) #graph sqrt(t)

legend('topleft', c("Kou","Square Root") , 
       lty=1, col=c("blue", "red"), bty='n', cex=.75) #legend
```
