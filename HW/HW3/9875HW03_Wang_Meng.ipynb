#### 1. (12 points) 

Consider the Heston model:

$$
\beas
\frac{dS_t}{S_t}&=&\sqrt{v_t}\,dZ\\
dv_t&=&-\lambda\,(v_t-\bar v)\,dt+\eta\,\sqrt{v_t}\,dW
\eeas
$$

with $\EE[dZ\,dW]=\rho\,dt$.  

(a)

Apply Itô's Lemma to derive the following expression for $\EE[S_T\,v_T]$:

<a name="eq:covsv"></a>(1)
$$
\begin{equation}
\EE[S_T\,v_T]=S_0\,v_0\,e^{-(\lambda-\rho\,\eta)\,T}+\lambda\,S_0\,\bar v\,\int_0^T\,e^{-(\lambda-\rho\,\eta)\,(T-t)}\,dt.
\end{equation}
$$


(b)

Integrate equation [(1)](#eq:covsv) explicitly to obtain a closed-form expression for the covariance $\Cov[S_T,v_T]$; for simplicity, assume the initial condition $v_0=\bar v$ so that $\hat v_t:=\EE[v_t]=\bar v$.

(c)
* What is the limit of your final expression as $\tau \to 0$?
* What is the covariance when $\rho=0$?
* What happens as $\lambda \to \infty$?

(d)

Explain these results to a non-specialist  *e.g.* your boss who is a trader.

### Answer
(a)
<br/>Apply product rule,
\begin{eqnarray}
d(S_tv_t)&=&v_tdS_t+S_tdv_t+dS_tdv_t\\
&=&v_t^{3/2}S_tdZ-\lambda(v_t-\bar v)S_tdt+\eta\sqrt{v_t}S_tdW+\eta v_tS_t\rho dt\\
&＝&v_t^{3/2}S_tdZ+\eta\sqrt{v_t}S_tdW+(\eta\rho-\lambda)S_tv_tdt+\lambda \bar v S_tdt
\end{eqnarray}
Integrate both sides,
$$
S_Tv_T-S_0v_0=\int_0^T v_T^{3/2}S_tdZ+\int_0^T \eta\sqrt{v_t}S_tdW+\int_0^T (\eta\rho-\lambda)S_tv_tdt+\int_0^T \lambda \bar v S_t dt
$$
Take expectations,
$$
\E[S_Tv_T]=S_0v_0+(\eta\rho-\lambda)\int_0^T \E[S_tv_t] dt+\lambda \bar v\int_0^T\E[S_t]dt
$$
Differeniate w.r.t. T,
$$
\frac{d\E[S_Tv_T]}{dT}=(\eta\rho-\lambda)E[S_Tv_T]+\lambda \bar v \E[S_T]
$$
The dynamic of $S_t$ shows that it is a martingale, thus, $\E[S_T]=S_0$
$$
\frac{d\E[S_Tv_T]}{dT}=(\eta\rho-\lambda)E[S_Tv_T]+\lambda \bar v S_0
$$
Solve the above ODE, we can get
$$
\E[S_Tv_T]=Ce^{(\eta\rho-\lambda)T}-\frac{\lambda \bar v S_0}{\eta\rho-\lambda}
$$
Plug into the initial condition, $\E[S_0v_0]=S_0v_0$, we can get $C=S_0v_0+\frac{\lambda \bar v S_0}{\eta\rho-\lambda}$
<br/>Finally,
$$
\E[S_Tv_T]=S_0v_0e^{(\eta\rho-\lambda)T}-\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right)
$$
which is exactly equation (1)

(b)
$$
\Cov[S_T,v_t]=E[S_Tv_T]-E[S_T]E[v_T]=S_0\bar ve^{(\eta\rho-\lambda)T}-\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right)-S_0\bar v=\frac{\eta\rho S_0\bar v}{\lambda-\eta\rho}\left(1-e^{-(\lambda-\eta\rho)T}\right)
$$

(c)
<br/> As $T$ goes to 0, covariance goes to 0.
<br/> As $\rho$ goes to 0, covariance goes to 0.
<br/> As $\lambda$ goes to 0, covariance goes to 0.

(d)
<br/>When $T$ goes to 0, $S_T$ and $v_T$ are very close to $S_0$ and $v_0$, so they are almost determined.
<br/>When $\rho$ goes to 0, the two random factors driving $S_t$ and $v_t$ are uncorrelated, which makes the covariance between $S_T$ and $v_T$ converges to 0.
<br/>When $\lambda$ goes to 0, the speed of mean-reverting of $v_t$ is infinity, $v_t$ is almost a constant, and no longer correlated with $S_t$.

#### 3.  (6 points.) 

Code a function in R to return the Heston characteristic function $\phi_T(u)$.  Verify that $\phi_T(0)=1$ and $\phi_T(-i)=1$.  Why must these latter two identities always hold (under the risk neutral measure)?

### Answer

```r
### Heston characteristic function
Heston_charac<-function(params){
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar; 
    v <- params$v;
    
    res <- function(u,t=1){
        al <- -u*u/2 - 1i*u/2;
        bet <- lambda - rho*eta*1i*u;
        gam <- eta^2/2;
        d <- sqrt(bet*bet - 4*al*gam);
        rp <- (bet + d)/(2*gam);
        rm <- (bet - d)/(2*gam);
        g <- rm / rp;
        D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
        C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) )
        return(exp(C*vbar+D*v))
    }
    return(res);
}
cat("value at 0: ",Heston_charac(paramsBCC)(0)," ")
cat("value at -i: ",Heston_charac(paramsBCC)(-1i))
```

$$
\phi_T(u)=\E\left[e^{iux_T}|x_t=0\right]
$$
$$
\phi_T(0)=\E\left[1|x_t=0\right]=1
$$
$$
\phi_T(-i)=\E\left[e^{x_T}|x_t=0\right]=\E\left[\left.\frac{S_T}{K}\right|S_0=K\right]=1
$$
The last equality comes from that under risk-neutral measure, $S_t$ is a martingale.

### Answer
(a) & (b)

```r
### probability density function
Heston_prob<-function(params){
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(x,t=1){
        f1<-function(u){
            al <- -u*u/2 - 1i*u/2;
            bet <- lambda - rho*eta*1i*u;
            gam <- eta^2/2;
            d <- sqrt(bet*bet - 4*al*gam);
            rp <- (bet + d)/(2*gam);
            rm <- (bet - d)/(2*gam);
            g <- rm / rp;
            D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
            C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) )
            return(Re(exp(C*vbar+D*v-1i*u*x)))
        } 
        p<-1/pi*integrate(function(u){f1(u)},lower=0,upper=Inf)$value;
        return(p)
    }
    return(res);
}
### t=1
yrange <- c(0,5)
prob<-function(x){sapply(x,function(x){Heston_prob(paramsBCC)(x)})}
curve(prob(x),from=-1.5,to=1,col="red",ylim=yrange,xlab="x",ylab="p(x,1)")

### Superimpose 5 dashed curves with eta changed
subH <- function(incr){
    tmp <- paramsBCC
    tmp$eta <- tmp$eta+incr
    return(tmp)
}

for (incr in seq(0.1,0.5,0.1))
    {
        for (i in 1:100000){tmp=i}
        par(new=T)
        prob <- function(x){sapply(x,function(x){Heston_prob(subH(incr))(x)})
    }
    curve(prob(x),from=-1.5,to=1,col="red",lty=2,ylim=yrange,ylab=NA,xlab=NA)
}
```
