#### Question 1 Solution

##### (a)

Apply Ito's lemma to $f(x)=S_tV_t$, we get

$$
\begin{eqnarray}
d(S_t v_t) 
&=& S_t d v_t + v_t d S_t + d [S, v]_t\\
&=& S_t (-\lambda (v_t-\bar v) dt+\eta \sqrt{v_t}dW) + V_t S_t \sqrt{v_t} dZ + \rho \eta S_t v_t dt\\
&=& S_t \lambda \bar v dt - S_t \lambda v_t dt + S_t \eta \sqrt{v_t} dW + V_t S_t \sqrt{v_t} dZ + \rho \eta S_t v_t dt\\
\end{eqnarray}
$$
Which is 
$$
\begin{eqnarray}
d(S_t v_t) = S_t \lambda \bar v dt - S_t \lambda v_t dt + S_t \eta \sqrt{v_t} dW + V_t S_t \sqrt{v_t} dZ + \rho \eta S_t v_t dt
\end{eqnarray}
$$

Take expectation of both sides, we have
$$
\begin{eqnarray}
d(\EE [S_t v_t]) = S_0 \lambda \bar v dt + (\rho \eta - \lambda) \EE [S_t v_t] dt
\end{eqnarray}
$$
since $\EE [S_t] = S_0$, $\EE [dW] = 0$, $\EE [dZ] = 0$.

Solve the ODE with $\EE [S_t v_t]$ as variable, we get
$$
E[S_T v_T] = S_0 v_0 e^{-(\lambda-\rho \eta) T}+\lambda S_0 \bar v \int_0^T e^{-(\lambda-\rho \eta) (T-t)} dt.
$$


#####(b)
$$
\begin{eqnarray}
Cov[S_T, v_T] &=& \EE[S_T v_T] - \EE [S_T]\EE[v_T] \\
&=& S_0 v_0 e^{-(\lambda-\rho \eta) T}+\lambda S_0 \bar v \int_0^T e^{-(\lambda-\rho \eta) (T-t)} dt - S_0 \bar v \\
&=& S_0 v_0 e^{-(\lambda-\rho \eta) T}+ \frac{\lambda S_0 \bar v(1-e^{-(\lambda - \rho\eta)T}）}{\lambda - \rho\eta} - S_0 \bar v
\end{eqnarray}
$$

#####（c)
As $T \to 0$, 

$$
\begin{eqnarray}
\lim_{T \to 0} {S_0 v_0 e^{-(\lambda-\rho \eta) T}+ \frac{\lambda S_0 \bar v(1-e^{-(\lambda - \rho\eta)T}）}{\lambda - \rho\eta} - S_0 \bar v}
&=&S_0 \bar v - S_0 \bar v\\
&=& 0
\end{eqnarray}
$$

When $\rho = 0$

$$
\begin{eqnarray}
Cov [S_T, v_T] &=& S_0 v_0 e^{-\lambda T}+ \frac{\lambda S_0 \bar v(1-e^{-\lambda T}）}{\lambda } - S_0 \bar v \\
&=& 0
\end{eqnarray}
$$

When $\lambda \to \infty$

$$
\begin{eqnarray}
\lim_{T \to 0} {S_0 v_0 e^{-(\lambda-\rho \eta) T}+ \frac{\lambda S_0 \bar v(1-e^{-(\lambda - \rho\eta)T}）}{\lambda - \rho\eta} - S_0 \bar v} 
&=& 0 + S_0 \bar v - S_0 \bar v \\
&=& 0
\end{eqnarray}
$$

#####(d)

$T \to 0$ is equivalent to say the option expires immediately,  $S_T$ and $v_T$ are just their current value $S_0$ and $v_0$, those are all known value, therefore, their covariance is 0.

When $\rho = 0$ which means the randomness of $v_t$ and $S_t$ are uncorrelated, therefore, their terminal value $v_T$ and $S_T$ should also be uncorrelated.

$\lambda \to \infty$ means that it takes no time for $v_t$ to mean revert. In other words, $v_t \equiv \bar v$, obviously, $ Cov[S_T, \bar v]=0$.

####3.  (6 points.) 

Code a function in R to return the Heston characteristic function $\phi_T(u)$.  Verify that $\phi_T(0)=1$ and $\phi_T(-i)=1$.  Why must these latter two identities always hold (under the risk neutral measure)?

```r
# cf of Heston model
CFHeston <- function(params){
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(u,t){
            j = 0
            al <- -u*u/2 - 1i*u/2 + 1i*j*u;
            bet <- lambda - rho*eta*1i*u-rho*eta*j;
            gam <- eta^2/2;
            d <- sqrt(bet*bet - 4*al*gam);
            rp <- (bet + d)/(2*gam);
            rm <- (bet - d)/(2*gam);
            g <- rm / rp;
            D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
            C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) );
            return(exp(C*vbar + D*v));
    }
    return(res);
}

subBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)
t <- seq(0.01, 10, length=100)

CF <- function(u){sapply(t,function(x){ CFHeston(subBCC)(u, t)})}
```

```r
# check \phi (0) and \phi(-i)
unique(as.vector(CF(0)))
unique(as.vector(CF(-1i)))
```

```r
PDFHeston <- function(params){  
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(k,t){
        fj <- function(u, t, j){

            al <- -u*u/2 - 1i*u/2 + 1i*j*u;
            bet <- lambda - rho*eta*1i*u-rho*eta*j;
            gam <- eta^2/2;
            d <- sqrt(bet*bet - 4*al*gam);
            rp <- (bet + d)/(2*gam);
            rm <- (bet - d)/(2*gam);
            g <- rm / rp;
            D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
            C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) );
            return(exp(C*vbar + D*v));}
        
        x <- -k;

        p <- 1/pi*integrate(function(u){Re(fj(u,t,0)*exp(-1i*u*x))},lower=0,upper=Inf)$value;
        
        return(p);
    }
    return(res);
}

subBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)

# for part (a)

tau <- 1
pdf <- function(x){sapply(x,function(x){PDFHeston(subBCC)(x, tau)})}

curve(pdf(x),from=-.5,to=.5,col="red",ylab="PDF",xlab="log strike")
```

(a)  Denote the Euler discretization of the Heston variance process by $\hat v$ and let $
\Delta \hat v_t = \hat v_{t+\Delta}- \hat v_t

(c) Using your results from (a) and (b), compute expressions for the discretization errors in expecation and variance to second order in $\Delta$.  Verify that the Euler discretization scheme has an $\cO(\Delta^2)$ bias, consistent with order one weak convergence.

#### Solution
#####(a)
We know that

$$
dv_t =-\lambda\,(v_t-\bar v)\,dt+\eta\,\sqrt{v_t}\,dW
$$

Therefore

$$
\Delta \hat v_t = -\lambda (v_t - \bar v) \Delta + \eta \sqrt v_t \sqrt \Delta Z
$$

Where $Z$ is a standard normal random variable. Therefore

$$
\begin{eqnarray}
\mathbb E [\Delta \hat v_t \mid v_t] 
&=& \mathbb E [-\lambda (v_t - \bar v) \Delta + \eta \sqrt v_t \sqrt \Delta Z \mid v_t] \\
&=& - \lambda (v_t - \bar v) \Delta\\
Var[\Delta \hat v_t \mid v_t] &=& \mathbb E [(\Delta \hat v_t - \mathbb E[\Delta \hat v_t \mid v_t])^2\mid v_t] \\
&=& \mathbb E[(\eta \sqrt v_t \sqrt \Delta Z)^2\mid v_t]\\
&=& {\eta}^2
v_t \Delta
\end{eqnarray}
$$

#####(b)

$$
\begin{eqnarray}
\mathbb E[v_{t+\Delta}|v_t]
&=& \hat v + (v_t - \hat v)e^{-\lambda \Delta}\\
&=& \hat v + (v_t - \hat v)(1 - \lambda \Delta + \frac{\lambda ^2}{2} \Delta ^2) + \mathcal O(\lambda ^3) \\
Var[v_{t+\Delta}|v_t] &=& \frac{\eta ^2}{\lambda}(1-e^{-\lambda \Delta})[e^{-\lambda \Delta}v_t + \frac 1 2 (1 - e^{-\lambda \Delta}\bar v]\\
&=& v_t \eta ^2 \Delta + \frac{\lambda \eta ^2 (\bar v - v_t)}{2} \Delta ^2 + \mathcal O(\Delta ^3)
\end{eqnarray}
$$

#####(c)

From above (a) and (b) we have
$$
\begin{eqnarray}
\mathbb E[v_{t+\Delta}|v_t] - \mathbb E [\Delta \hat v_t \mid v_t]
&=& \frac{\lambda ^2(v_t - \hat v)}{2} \Delta ^2 + \mathcal O(\lambda ^3) \\
Var[v_{t+\Delta}|v_t] - Var[\Delta \hat v_t \mid v_t]
&=& \frac{\lambda \eta ^2 (\bar v - v_t)}{2} \Delta ^2 + \mathcal O(\Delta ^3)
\end{eqnarray}
$$
Both have order two convergence.
