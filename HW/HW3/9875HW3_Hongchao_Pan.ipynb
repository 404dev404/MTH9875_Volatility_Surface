### Answer of Q1

#### Answer of (a): Ito's Lemma

* Apply Ito's formula to $(S_tv_t)$ we have:

\begin{eqnarray*}
d(S_tv_t) &=& v_tdS_t +S_tdv_t +dS_tdv_t \\
&=&v_t^{3/2}S_tdZ -\lambda S_t(v_t-\bar{v})dt+\eta S_t v_t^{1/2}dW+\eta v_tS_t\rho dt\\
&=& v_t^{3/2}S_tdZ +(\rho \eta - \lambda)S_tv_tdt+\lambda \bar{v}S_tdt +\eta S_tv_t^{1/2}dW
\end{eqnarray*}

* Integrate on both sides, we have:

\begin{eqnarray*}
S_Tv_T=S_0v_0+\int_0^T v_t^{3/2}S_tdZ +\int_0^T (\rho \eta - \lambda)S_tv_tdt +\int_0^T \lambda \bar{v}S_tdt +\int_0^T \eta S_tv_t^{1/2}dW
\end{eqnarray*}

* Take the expectation on both sides, we have:

$$
\mathbb{E}\left[S_Tv_T\right]=S_0v_0 +(\rho \eta -\lambda)\int_0^T\mathbb{E}[S_tv_t]dt + \lambda \bar{v}\int_0^T\mathbb{E}[S_t]dt
$$

* Since there is no drift term in the dynamics of $S_t$, then $\mathbb{E}[S_t]=S_0$


* Denote $y=\mathbb{E}[S_Tv_T]$, then take derivative on both sides w.r.t. to $T$, we have:

$$
y'-(\rho \eta-\lambda)y=\lambda \bar{v}S_0
$$

* Solving above ODE, we have:

$$
\mathbb{E}[S_Tv_T]=Ce^{(\rho \eta-\lambda)T}-\frac{\lambda \bar v S_0}{\rho \eta-\lambda}
$$

* Apply initial condition with $T=0$, we have $C=S_0v_0+\frac{\lambda \bar v S_0}{\rho\eta-\lambda}$

* Therefore,
\begin{eqnarray*}
\mathbb{E}[S_Tv_T] &=& S_0v_0 e^{-(\lambda-\rho \eta)T} +\frac{\lambda \bar v S_0}{\rho \eta-\lambda} e^{-(\lambda-\rho\eta)T} - \frac{\lambda \bar v S_0}{\rho \eta-\lambda}\\
&=& S_0v_0 e^{-(\lambda-\rho \eta)T} -\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right)
\end{eqnarray*}

* As
$$
\lambda\,S_0\,\bar v\,\int_0^T\,e^{-(\lambda-\rho\,\eta)\,(T-t)}\,dt = -\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right)
$$


* Thus, equation (1) was proofed.

#### answer of (b): $\Cov[S_T,v_T]$

* As the dynamic of $S_t$ does not have drift term, then $\EE[S_T]=S_0$


* With $\EE[v_t]=\bar{v}$, we have:

\begin{eqnarray*}
\Cov[S_T,v_T]&=&\EE{S_Tv_T}-\EE{S_T}\EE{v_T} \\
&=& S_0v_0 e^{-(\lambda-\rho \eta)T} -\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right) - S_0\bar{v}\\
&=& -S_0\bar{v}\left(1-e^{(\eta\rho-\lambda)T}\right) -\frac{\lambda \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right) \\
&=& \frac{\rho\eta \bar v S_0}{\eta\rho-\lambda}\left(1-e^{(\eta\rho-\lambda)T}\right)
\end{eqnarray*}

#### Answer of (c): Extreme conditions

* If $T \rightarrow 0$. Thus, from the formula we have $\Cov[S_T,V_T] \rightarrow 0$.


* If $\rho \rightarrow 0$, then from the formula we have $\Cov[S_T,V_T] \rightarrow 0$.


* If $\lambda \rightarrow \infty$, then from the formula we have $\Cov[S_T,V_T] \rightarrow 0$.


#### Answer of (d): General explanation

* If $T \rightarrow 0$. This means that the option expires immediately, therefore $S_T$ and $v_T$ are determinstic, i.e., their initial value, $S_0$ and $v_0$. Thus, the covariance of $S_T$ and $v_T$ is 0.

* If $\rho \rightarrow 0$. This means $S_t$ and $v_t$ are uncorrelated. At maturity, $S_T$ and $v_T$ should also be uncorrelated. Therefore the covariance of $S_T$ and $v_T$ is 0.

* If $\lambda \rightarrow \infty$. This means the mean reverting rate of $v_t$ is infinity, which means the mean reverting time of $v_t$ to $\bar{v}$ is 0. This implies that $v_t \equiv \bar{v}$ all the time. Therefore, $\Cov[S_T,v_T] = \Cov[S_T,\bar{v}]=0$

### Answer of Q3

```r
# Define a function to compute Heston characteristic function
CF_Heston<-function(params){
    # Get the parameters from params, assume paramsBCC
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(u,t){
        al <- -u*u/2 - 1i*u/2;
        bet <- lambda - rho*eta*1i*u;
        gam <- eta^2/2;
        d <- sqrt(bet*bet - 4*al*gam);
        rp <- (bet + d)/(2*gam);
        rm <- (bet - d)/(2*gam);
        g <- rm / rp;
        D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
        # By using Jim Gatheral's big C
        C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) )
        return(exp(C*vbar+D*v))
    }
    return(res);
}
# Get BCC parameters
paramsBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)
t <- 1 # Assume maturity is 1 year
char_Heston <- function(u){sapply(t,function(x){ CF_Heston(paramsBCC)(u, t)})}

# Check the series of t
t2 <- seq(0.001,10,length=10)
char_Heston2 <- function(u){sapply(t2,function(x){ CF_Heston(paramsBCC)(u, t2)})}
```

#### Answer of (a): plot density function

```r
# Code a function for equation (17) in the lecture 3

pdf_Heston<-function(params){
    # Assume BCC parameters
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v <- params$v;
    
    res <- function(k,t){
        fj <- function(u, t, j){

            al <- -u*u/2 - 1i*u/2 + 1i*j*u;
            bet <- lambda - rho*eta*1i*u-rho*eta*j;
            gam <- eta^2/2;
            d <- sqrt(bet*bet - 4*al*gam);
            rp <- (bet + d)/(2*gam);
            rm <- (bet - d)/(2*gam);
            g <- rm / rp;
            D <- rm * (1 - exp(-d*t))/ (1 - g*exp(-d*t));
            # By using Jim Gatheral's big C
            C <- lambda * (rm * t - 2/eta^2 * log( (1 - g*exp(-(d*t)))/(1 - g) ) );
            
            return(exp(C*vbar + D*v));
        }
        
        x <- -k;

        p <- 1/pi*integrate(function(u){Re(fj(u,t,0)*exp(-1i*u*x))},lower=0,upper=Inf)$value;
        
        return(p);
    }
    return(res);
}

# Get BCC parameters
paramsBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = .04)

# tau=1
t <- 1

pdf_Heston_cal <- function(x){sapply(x,function(x){pdf_Heston(paramsBCC)(x, t)})}
# Plot the density function
curve(pdf_Heston_cal(x),from=-.5,to=.5,col="blue",ylab="Heston probability",
      xlab="Log strike",main="Plot of Heston probability density function",lwd=3)
```

### Answer of Q5

#### Answer of (a): Conditional expectation and conditional variance

* From lecture #3, we know the dynamic of Heston variance process has following SDE:

$$
dv_t =-\lambda\,(v_t-\bar v)\,dt+\eta\,\sqrt{v_t}\,dW
$$

* Therefore, the discretized form is:

$$
\Delta \hat v_t = -\lambda (v_t - \bar v) \Delta + \eta \sqrt v_t \sqrt \Delta Z
$$

* With $Z \sim N(0,1)$, we have:


\begin{eqnarray*}
\mathbb E [\Delta \hat v_t \mid v_t] 
&=& \mathbb E [-\lambda (v_t - \bar v) \Delta \mid v_t] +\mathbb E [\eta \sqrt v_t \sqrt \Delta Z \mid v_t] \\
&=& - \lambda (v_t - \bar v) \Delta\\
Var[\Delta \hat v_t \mid v_t] &=& \mathbb E [(\Delta \hat v_t - \mathbb E[\Delta \hat v_t \mid v_t])^2\mid v_t] \\
&=& \mathbb E[(\eta \sqrt v_t \sqrt \Delta Z)^2\mid v_t]\\
&=& {\eta}^2
v_t \Delta
\end{eqnarray*}


#### Answer of (b): Exact erpression

* From lecture notes, we have:

\begin{eqnarray*}
\mathbb E[v_{t+\Delta}|v_t]
&=& \bar v + (v_t - \bar v)e^{-\lambda \Delta}\\
\end{eqnarray*}


* Using Taylor expansion, we have:

\begin{eqnarray*}
\mathbb E[v_{t+\Delta}|v_t]
&=& \bar v + (v_t - \bar v)e^{-\lambda \Delta}\\
&\approx& \bar v + (v_t - \bar v)(1 - \lambda \Delta + \frac{\lambda ^2}{2} \Delta ^2) + \mathcal O(\Delta ^3)
\end{eqnarray*}

* Similar for conditional variance, we have: 

\begin{eqnarray*}
\Var[v_{t+\Delta}|v_t] &=& \frac{\eta ^2}{\lambda}(1-e^{-\lambda \Delta})[e^{-\lambda \Delta}v_t + \frac 1 2 (1 - e^{-\lambda \Delta}\bar v]\\
&=& v_t \eta ^2 \Delta + \frac{\lambda \eta ^2 (\bar v - v_t)}{2} \Delta ^2 + \mathcal O(\Delta ^3)
\end{eqnarray*}

#### Answer of (c): discretization errors

* From above (a) and (b) we have

\begin{eqnarray*}
\mathbb E[v_{t+\Delta}|v_t] - \mathbb E [\Delta \hat v_t \mid v_t]
&=& \bar{v}+ \frac{\lambda ^2(v_t - \bar v)}{2} \Delta ^2 + \mathcal O(\Delta ^3) \\
Var[v_{t+\Delta}|v_t] - Var[\Delta \hat v_t \mid v_t]
&=& \frac{\lambda \eta ^2 (\bar v - v_t)}{2} \Delta ^2 + \mathcal O(\Delta ^3)
\end{eqnarray*}


* From the erros above, we can conclude that the Euler discretization scheme has an $\cO(\Delta^2)$ bias, consistent with order one weak convergence.
