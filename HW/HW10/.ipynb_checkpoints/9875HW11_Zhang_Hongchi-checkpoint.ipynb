Assume a stock process of the form

$$
\frac{dS_t}{S_t}=\sigma_t\,dZ_t
$$

where $\sigma_t$ may be stochastic.

A continuous time version of the payoff of a gamma swap is
$
\int_0^T\,S_t\,\sigma_t^2\,dt
$.

(a)

Compute $d(S_t\,\log S_t)$ using It√¥'s Lemma.

<font color = 'blue'>

$$
Let \,\,f(x) = x\,ln(x), f_x = ln(x) + 1, f_{xx} = \frac{1}{x}
$$

<font color = 'blue'>

Apply Ito's lemma and we get the following equation (1)

\begin{eqnarray}
d(S_t\,\log S_t)&=& (1+log S_t) dS_t + \frac12 \frac{1}{S_t} (dS_t)^2 \\
&=& (1+log S_t) dS_t + \frac12\,S_t\,\sigma_t^2
\end{eqnarray}

(b)
Deduce the semi-static model-free hedging strategy involving a static hedge in European options and continuous rebalancing in the stock. 

<font color = 'blue'>

Integrate both sides of equation (1) we get:

\begin{eqnarray}
\int_0^T\,S_t\,\sigma_t^2\,dt = 2S_T log S_T - 2S_0 log S_0  + \int_0^T -2(1 + log S_t)dS_t
\end{eqnarray}



<font color = 'blue'>
So to hedge a gamma swap, we need to 

1) Static replication of $2 S_T log S_T$ using European Options

2) Dynamically keeping a short position of $2 (1 + log S_t)$ shares in stock

3) Short $2 S_0 log S_0$ in cash

(c) 

Using the spanning formula, derive an expression for the weights of the options in the static hedge.

<font color = 'blue'>

The spanning formula is 

Then, with $F=\E[S_T]$, 

<a name="eq:generalizedpayoff"></a>(2)
\begin{eqnarray*}
 \mathbb{E} \left[ g(S_T)\right]
 =g(F)\,+\,\int _0 ^F \, dK \, \tilde
 P(K)\,g''(K)\,+\,\int _F ^\infty \, dK \,
  \tilde C(K)\,g''(K)
\end{eqnarray*}

<font color = 'blue'>

Let g(x) = 2 x ln(x)

$$
    g''(x) = \frac{2}{x}\\
    g''(K) = \frac{2}{K}
$$

Therefore,the weights of the options in the static hedge is $\frac2K$

## Demonstrate replication path-by-path

### 2. (10 points) 

(a)

Amend (for example) the stochastic volatility and local volatility Monte Carlo codes (*HestonMCxmax* and *HestonLvMCxmax* respectively) from Lecture 10, with the same HN parameters:

$$
\begin{eqnarray*} v&=& 0.04\\ \bar v &=& 0.04\\ \lambda&=&10\\
\eta&=&1\\\rho&=&-1
\end{eqnarray*}
$$

to demonstrate that the 
variance swap hedging strategy outlined on page 136 of *The
Volatility Surface* replicates the variance swap payoff path-by-path independently of the dynamical assumptions.  Assume a 1-year variance swap, use 252 time steps and 1,000 paths.  

```r
paramsHN <- list(lambda = 10,rho = -1,eta = 1,vbar = 0.04,v = 0.04)

evolveAlfonsiF <- function(v,x,dt,W1,W2,L){
  
  eldt2 <- exp(-lambda*dt/2);
  
  #Variance process 
  vbarp <- vbar - eta^2/(4*lambda);
  psi <- (1-eldt2)/lambda;
  v1 <- v*eldt2+lambda*vbarp*psi;
  v2 <- (v1 > 0) * v1; # Take v2 = 0 if v1<0, else v2=v1
  par <- sqrt(v2) + eta/2 * sqrt(dt)*W2;
  vf <- eldt2*par*par +lambda*vbarp*psi + v1 - v2;
  # Full truncation
  
  # Log-stock process (Andersen equation (33))
  vvf <- (v+vf > 0) * (v+vf);
  dw <- vvf/2*dt;
  x <- x  - dw/2 + rho2m1*sqrt(dw)*W1 +
    rho/eta*(lambda*dw + vf-v -lambda*vbar*dt) ;
  # Impose martingale constraint
  x <- x - log(mean(exp(x)));
  v <- vf;
  return(cbind(x,v));
}

HestonMCSv_vs <- function(params){

  is.even <- function(j){as.logical((j+1) %% 2)}
  
  res <- function(S0, T, AK, N, m, evolve)
  {
    
    lambda <<- params$lambda 
    rho <<- params$rho 
    eta <<- params$eta 
    vbar <<- params$vbar 
    v0 <<- params$v 
    
    n <- m
    sqrt2 <- sqrt(2) 
    rho2m1 <<- sqrt(1-rho*rho) 
      
    vbarp <<- vbar - eta^2/(4*lambda) 
    
    negCount <- 0 
    
    # We use a vertical array, one element per M.C. path
    x <- numeric(N);  v <- rep(1,N)*v0  
    stockpnl <-numeric(N)
    stockpos <-  1/ exp(x) 
    Totalvar <-numeric(N)   
    W1m <- numeric(N) 
    W2m <- numeric(N)  
    
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
      # Two sets of correlated normal random vars.
      
      W1 <- rnorm(N) 
      W2 <- rnorm(N) 
      W1 <- W1 - mean(W1)
      W1 <- W1/sd(W1) 
      W2 <- W2 - mean(W2)
      W2 <- W2/sd(W2) 
      # Now W1 and W2 are forced to have mean=0 and sd=1
      
      W2p <- W2 - cor(W1,W2)*W1  # Eliminate actual correlation
      W2p <- W2p - mean(W2p)  
      W2 <- W2p/sd(W2p)  
      # Now W1 and W2 have mean=0, sd=1 and correlation=0
      
      L <- rbinom(N, size=1, prob=1/2)  # Bernoulli rv for NV step
         
      res <- evolve(v,x,T/n,W1,W2,L) 
      xnew <- res[,1] 
      vnew <- res[,2]
    
      #Update stock hedging pnl
      stockpnl = stockpnl + stockpos * (exp(xnew) - exp(x))
      #Update Stock position
      stockpos = 1/ exp(xnew)   
      #Udpate Total Var
      Totalvar = Totalvar + (v + vnew )/2.* T / n /2
      #---------------------------------------------------------------------------------------------------------------  
      #Roll forward x and v
      x <- xnew 
      v <- vnew 
    }
    
    return(data.frame(x,v,stockpnl,Totalvar)) 
    
  }
  return(res) 
}

# Monte Carlo  - Local Volatility
HestonLv_vs <- function(params){

  is.even <- function(j){as.logical((j+1) %% 2)}
  
  res <- function(S0, T, N, m)
  {
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v0 <- params$v;
    
    lambdap <- lambda+eta/2;
    vbarp <- vbar*lambda/lambdap;
    
    n <- m;
    dt <- T/n;
    dtm <- dt*2;
    
    sqrtdt <- sqrt(dt);
    sqrtdtm <- sqrt(dtm);
    sqrt2 <- sqrt(2);
    
    # We use a vertical array, one element per M.C. path
    x <- numeric(N); vloc <- rep(v0,N); 
    stockpnl <-numeric(N)
    stockpos <-  1/ exp(x) 
    Totalvar <-numeric(N)   
    
    # Loop over timesteps
    for (i in 1:n)
    {
      W1 <- rnorm(N);
      W1 <- W1 - mean(W1); W1 <- W1/sd(W1);
      # Now W1 is forced to have mean=0 and sd=1  
      # Add code for subgrid
     # W1m <- W1m + W1/sqrt2; # N(0,1) rv's for subgrid
      # Compute local variance from formula 4.1 of TVS
      ti <- i*dt;
      elti <- exp(-lambdap*ti);	
      vlocf <- (v0-vbarp)*elti+vbarp-eta*x*(1-elti)/(lambdap*ti);
      vlocf <- (vlocf>=0)*vlocf;
      
      # Approximate integrated variance
      w <- (vlocf +vloc)/2*dt;
      xnew <- x - w/2 + sqrt(w)*W1;
      stockpnl = stockpnl + stockpos * (exp(xnew) - exp(x))
      #Update Stock position
      stockpos = 1/ exp(xnew)   
      #Udpate Total Var
      Totalvar = Totalvar + w/2 
      #--------------------------------------------------------------------------------------------------------------
      x <- xnew;
      vloc <- vlocf;
    }
    return(data.frame(x,vloc,stockpnl,Totalvar));
  }
  return(res);
}

```

```r
res_heston = HestonMCSv_vs(paramsHN)(S0=1, T=1, AK=1, N=1000, m=252, evolve=evolveAlfonsiF)
res_lv = HestonLv_vs(paramsHN)(S0=1, T=1, N=1000, m=252)
```

```r
#Static Replication of log contracts using European Options 
option_vs <- function(x){
  dK = 0.01
  Ks = seq(0.001, 1000, dK)  
  S = exp(x) 
  weight_call = 2*(Ks > 1)/(Ks^2)#OTM Call
  weight_put = 2*(Ks < 1)/(Ks^2)#OTM Put
  res = sapply(S, function(spot){
    return(sum((spot>=Ks)*(spot-Ks)*weight_call*dK) + sum((spot<Ks)*(Ks-spot)*weight_put*dK)) 
  })
  return(-res)
}
```

<font color = 'blue'>

In our case, F = $S_0$ = 1, $g(S_T) = log(S_T)$

\begin{eqnarray*}
 g(S_T)&=&\int_0^F\,g''(K)\,(K-S_T)^+\,dK + \int_F^\infty\,g''(K)\,(S_T-K)^+\,dK\nonumber\\
 &&+ g(F) - g'(F)\,\left[ (F-S_T)^+ - (S_T-F)^+ \right]\nonumber\\
 &=&\int_0^F\,g''(K)\,(K-S_T)^+\,dK + \int_F^\infty\,g''(K)\,(S_T-K)^+\,dK\nonumber\\
 &&+ g(F)  + g'(F)\,(S_T-F).\\
 &=&\int_0^1\,g''(K)\,(K-S_T)^+\,dK + \int_1^\infty\,g''(K)\,(S_T-K)^+\,dK\nonumber\\
 &&+ (S_T-1)
\end{eqnarray*}

We can replicate the log contract path by path by trading strip of European options plus one share of stock

```r
heston_error = res_heston$stockpnl*2 - res_heston$Totalvar*2 - option_vs(res_heston$x)-2*(exp(res_heston$x)-1)
lv_error = res_lv$stockpnl*2 - res_lv$Totalvar*2 - option_vs(res_lv$x)- 2*(exp(res_lv$x)-1)
```

### 3. (6 points) 

Repeat Problem 2 for a gamma swap with the static hedge you derived for the gamma swap in Problem 1.

\begin{eqnarray}
\int_0^T\,S_t\,\sigma_t^2\,dt = 2S_T log S_T - 2S_0 log S_0  + \int_0^T -2(1 + log S_t)dS_t
\end{eqnarray}


```r
#Define Monte Carlo Simulation for Gamma Swap
gammaswap_HestonMC <- function(params){

  is.even <- function(j){as.logical((j+1) %% 2)}
  
  res <- function(S0, T, AK, N, m, evolve)
  {
    
    lambda <<- params$lambda 
    rho <<- params$rho 
    eta <<- params$eta 
    vbar <<- params$vbar 
    v0 <<- params$v 
    
    n <- m
    sqrt2 <- sqrt(2) 
    rho2m1 <<- sqrt(1-rho*rho) 
      
    vbarp <<- vbar - eta^2/(4*lambda) 
    
    negCount <- 0 
    
    # We use a vertical array, one element per M.C. path
    x <- numeric(N);  v <- rep(1,N)*v0  
    stockpnl <-numeric(N)
    stockpos <-  -2*(1+x)
    Totalvar <-numeric(N)   
    W1m <- numeric(N) 
    W2m <- numeric(N)  
    dt = T/n
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
      # Two sets of correlated normal random vars.
      
      W1 <- rnorm(N) 
      W2 <- rnorm(N) 
      W1 <- W1 - mean(W1)
      W1 <- W1/sd(W1) 
      W2 <- W2 - mean(W2)
      W2 <- W2/sd(W2) 
      # Now W1 and W2 are forced to have mean=0 and sd=1
      
      W2p <- W2 - cor(W1,W2)*W1  # Eliminate actual correlation
      W2p <- W2p - mean(W2p)  
      W2 <- W2p/sd(W2p)  
      # Now W1 and W2 have mean=0, sd=1 and correlation=0
      
      L <- rbinom(N, size=1, prob=1/2)  # Bernoulli rv for NV step
         
      res <- evolve(v,x,T/n,W1,W2,L) 
      xnew <- res[,1] 
      vnew <- res[,2]
    
      #Update stock hedging pnl
      stockpnl = stockpnl + stockpos * (exp(xnew) - exp(x))
      #Update Stock position
      stockpos = -2*(1+xnew)   
      #Udpate Total Var
      Totalvar = Totalvar + (v*exp(x) + vnew*exp(xnew))/2.* dt
      #---------------------------------------------------------------------------------------------------------------  
      #Roll forward x and v
      x <- xnew 
      v <- vnew 
    }
    
    return(data.frame(x,v,stockpnl,Totalvar)) 
    
  }
  return(res) 
}

# Monte Carlo  - Local Volatility
gammaswap_LVMC <- function(params){

  is.even <- function(j){as.logical((j+1) %% 2)}
  
  res <- function(S0, T, N, m)
  {
    
    lambda <- params$lambda;
    rho <- params$rho;
    eta <- params$eta;
    vbar <- params$vbar;
    v0 <- params$v;
    
    lambdap <- lambda+eta/2;
    vbarp <- vbar*lambda/lambdap;
    
    n <- m;
    dt <- T/n;
    dtm <- dt*2;
    
    sqrtdt <- sqrt(dt);
    sqrtdtm <- sqrt(dtm);
    sqrt2 <- sqrt(2);
    
    # We use a vertical array, one element per M.C. path
    x <- numeric(N); vloc <- rep(v0,N); 
    stockpnl <-numeric(N)
    stockpos <--2*(1+x)  
    Totalvar <-numeric(N)   
    
    # Loop over timesteps
    for (i in 1:n)
    {
      W1 <- rnorm(N);
      W1 <- W1 - mean(W1); W1 <- W1/sd(W1);
      # Now W1 is forced to have mean=0 and sd=1  
      # Add code for subgrid
      # W1m <- W1m + W1/sqrt2; # N(0,1) rv's for subgrid
      # Compute local variance from formula 4.1 of TVS
      ti <- i*dt;
      elti <- exp(-lambdap*ti);	
      vlocf <- (v0-vbarp)*elti+vbarp-eta*x*(1-elti)/(lambdap*ti);
      vlocf <- (vlocf>=0)*vlocf;
      
      # Approximate integrated variance
      w <- (vlocf +vloc)/2*dt;
      xnew <- x - w/2 + sqrt(w)*W1;
      var = (vlocf*exp(xnew) + vloc*exp(x))/2*dt
      stockpnl = stockpnl + stockpos * (exp(xnew) - exp(x))
      #Update Stock position
      stockpos = -2*(1+xnew)   
      #Udpate Total Var
      Totalvar = Totalvar + var
      #--------------------------------------------------------------------------------------------------------------
      x <- xnew;
      vloc <- vlocf;
    }
    return(data.frame(x,vloc,stockpnl,Totalvar));
  }
  return(res);
}
```

We can replicate the log contract path by path by trading strip of European options plus one share of stock

```r
#Static Replication of log contracts using European Options 
option_gs <- function(x){
  dK = 0.01 
  Ks = seq(0.001, 1000, dK)  
  S = exp(x) 
  weight_call = 2*(Ks > 1)/Ks  #OTM Call
  weight_put = 2*(Ks < 1)/Ks   #OTM Put
  res = sapply(S, function(spot){
    return(sum((spot>=Ks)*(spot-Ks)*weight_call*dK) + sum((spot<Ks)*(Ks-spot)*weight_put*dK)) 
  })
  return(res)
}
```

```r
#Compute Monte Carlo Sims
gs_heston = gammaswap_HestonMC(paramsHN)(S0=1, T=1, AK=1, N=1000, m=252, evolve=evolveAlfonsiF)
gs_lv = gammaswap_LVMC(paramsHN)(S0=1, T=1, N=1000, m=252)
gs_heston_error = gs_heston$stockpnl - gs_heston$Totalvar + 2*option_gs(gs_heston$x)+2*(exp(gs_heston$x)-1)
gs_lv_error = gs_lv$stockpnl - gs_lv$Totalvar + 2*option_gs(gs_lv$x)+2*(exp(gs_lv$x)-1)
```

```r
#Calculate mean and standar deviation of hedging errors
print("Path by path Gamma Swap Hedging errors in Stochastic Volatility Model :")
paste('mean :',mean(gs_heston_error))
paste('standard deviation :',sd(gs_heston_error))
print("Path by Path Gamma Swap Hedging errors in Local Volatility Model :")
paste('mean :',mean(gs_lv_error))
paste('standard deviation :',sd(gs_lv_error))
```
