#### 1. (4 points)

Let $X_i$ be independent power-law distributed random variables, so that for large $x$,

$$
\Pr(X_i>x) \sim \frac{1}{x^{\zeta_{X_i}}} 
$$

 Prove the following properties of the tail exponents $\zeta_{X_i}$:

 (a) $$
\zeta_{X^\alpha}=\frac{\zeta_{X}}{\alpha}%\label{eq:PLpower}
$$

(b) $$
\zeta_{\alpha\,X}=\zeta_{X}
%\label{eq:PLprod}
$$



### Answers of Q1

#### Answer of (a)

* From lecture 1 and power-law fomula, we assume $\alpha>0$, then we have:

$$
\Pr(X^{\alpha}>x) \sim \frac{1}{x^{\zeta_{X^{\alpha}}}} 
$$

    and


$$
\Pr(X>x^{\frac{1}{\alpha}}) \sim \frac{1}{x^{\frac{1}{\alpha} \zeta_{X}}} 
$$

$\Rightarrow$

$$
\zeta_{X^\alpha}=\frac{\zeta_{X}}{\alpha}%\label{eq:PLpower}
$$



#### Answer of (b)

* Similar assumption in (a) that $\alpha>0$, then we have:

$$
\Pr(\alpha X>x) = \Pr\left(X>\frac{x}{\alpha}\right) \sim \frac{1}{\left(\frac{x}{\alpha}\right)^{\zeta_{X}}} =\frac{\alpha^{\zeta_X}}{x^{\zeta_X}} \sim \frac{1}{x^{\zeta_X}}
$$

    and
    
$$
\Pr({\alpha}X>x) \sim \frac{1}{x^{\zeta_{{\alpha}X}}} 
$$

    Then we have
    
$$
\zeta_{\alpha X}=\zeta_X
$$

### Answer of Q4

#### Answer of (a): SPX move

* Based on the properties of Student-t distribution and assume X has standard Student-t distribution with degree of freedom as 3. 

* Denote Y as the SPX in Q4 with $\mu=0, df=3, \sigma=0.01$, then we have $Y$ has the same distribution as $X\,\frac{0.01}{\sqrt{3}}$:

$$
Y \sim X\,\frac{0.01}{\sqrt{3}}
$$

* Therefore, $\Pr(Y>\alpha)=\Pr(X>100\sqrt{3}\alpha)$


* $\alpha=0.1$ 

```r
df <-3
alpha <- 0.1
p10 <- 2*(1-pt(100*sqrt(3)*alpha, df=df))
cat("The chance of a daily move of more than 10% is: ", round(p10,9),"\n")
```

* $\alpha=0.2$

```r
df <-3
alpha <- 0.2
p20 <- 2*(1-pt(100*sqrt(3)*alpha, df=df))
cat("The chance of a daily move of more than 20% is: ", round(p20,9),"\n")
```

* Years of move more than 10% or 20%. Assume 252 trading days per year.

```r
# 10%
year_10 <- 1/p10/252
#20%
year_20 <- 1/p20/252
cat("We should expect to wait ", round(year_10,2), " years to get daily moves over 10% ","\n")
cat("We should expect to wait ", round(year_20,2), " years to get daily moves over 20% ","\n")
```

### Answer of Q6

#### Answer of (a): R code

* The average log-return conditional on the log-return being non-negative.

```r
# Get the data
VIX <- spxVixData$Cl.VIX
SPX <- spxVixData$Cl.GSPC
# Check the data
tail(SPX)
```

```r
# Get the log return of VIX and SPX
log_VIX <- log(VIX/lag(VIX))
log_SPX <- log(SPX/lag(SPX))
# Drop the 1st data point (NA)
log_VIX <- log_VIX[!is.na(log_VIX)]
log_SPX <- log_SPX[!is.na(log_SPX)]
```

```r
# Compute the average of log-run conditional on log-return as non-negative
Avg_VIX_N <- mean(log_VIX[log_VIX>=0.0])
Avg_SPX_N <- mean(log_SPX[log_SPX>=0.0])

cat("The average of log-return of VIX conditional on the log-return being non-negative is: ",
   round(Avg_VIX_N,9),"\n")
cat("The average of log-return of SPX conditional on the log-return being non-negative is: ",
   round(Avg_SPX_N,9),"\n")
```

* The average log-return conditional on the log-return being negative.

```r
# Compute the average of log-run conditional on log-return as negative
Avg_VIX_Neg <- mean(log_VIX[log_VIX < 0.0])
Avg_SPX_Neg <- mean(log_SPX[log_SPX < 0.0])

cat("The average of log-return of VIX conditional on the log-return being negative is: ",
   round(Avg_VIX_Neg,9),"\n")
cat("The average of log-return of SPX conditional on the log-return being negative is: ",
   round(Avg_SPX_Neg,9),"\n")
```

#### Answer of (b): What proportion of all moves are up-moves? Down-moves?

```r
# Compute the proportion based on length of the dataset
# Up-moves
VIX_UP <- length(log_VIX[log_VIX > 0.0])/length(log_VIX)
SPX_UP <- length(log_SPX[log_SPX > 0.0])/length(log_SPX)

# Down-moves
VIX_DOWN <- length(log_VIX[log_VIX < 0.0])/length(log_VIX)
SPX_DOWN <- length(log_SPX[log_SPX < 0.0])/length(log_SPX)

cat("The propotion of VIX are up-moves is: ", round(VIX_UP,9),"\n")
cat("The propotion of SPX are up-moves is: ", round(SPX_UP,9),"\n")

cat("The propotion of VIX are down-moves is: ", round(VIX_DOWN,9),"\n")
cat("The propotion of SPX are down-moves is: ", round(SPX_DOWN,9),"\n")
```

#### answer of (c): Relate your findings to one of the Rama Cont stylized facts

* As we can see from (b) that the up-moves/down-moves of both VIX and SPX are close to 50%/50%. This observation is "Gain/loss asymmetry" in Rama Cont's stylized facts.

### Answer of Q7

#### Answer of (a):  download time series of DAX and VDAX

* Notes: Two DAX found in yahoo finance

* DAX: https://finance.yahoo.com/quote/DAX?p=DAX

* DAX(^GDAXI): https://finance.yahoo.com/quote/%5EGDAXI?ltr=1

* The 1st data point of DAX is 2014, but 1st data point of DAX(^GDAXI) is 1990. Therefore, chose DAX(^GDAXI) as my DAX

```r
# Customized from code in lecture
#Get DAX and VDAX data from Yahoo!
options("getSymbols.warning4.0"=FALSE)
getSymbols("^GDAXI",from="1927-01-01") # Creates the time series object DAX
getSymbols("^VDAX",from="1927-01-01") # Creates the time series object VDAX
```

```r
mm <- specifyModel(Cl(GDAXI)~Cl(VDAX))
DAX_VDAX_Data<-modelData(mm) #quantmod function automatically aligns data from two series
head(DAX_VDAX_Data)
```

#### Answer of (b): Regress VDAX log returns vs DAX log returns

```r
DAX <- DAX_VDAX_Data$Cl.GDAXI
VDAX <- DAX_VDAX_Data$Cl.VDAX
# Check the data
head(DAX)
```

```r
# Compute the log-return
log_DAX <- log(DAX/lag(DAX))
log_VDAX <- log(VDAX/lag(VDAX))
# Drop NA
log_DAX <- log_DAX[!is.na(log_DAX)]
log_VDAX <- log_VDAX[!is.na(log_VDAX)]
head(log_DAX)
```

```r
# Regression
reg_DAX <- lm(log_VDAX ~ log_DAX)
# Get the R2
R2 <- summary(reg_DAX)$r.squared
Adj_R2<-summary(reg_DAX)$adj.r.squared
# Get the correlation
corr <- cor(log_VDAX,log_DAX)
```

```r
cat("The R2 of the regression is ",round(R2,4),"\n")
cat("The Adjust R2 of the regression is ",round(Adj_R2,4),"\n")
cat("The correlation is: ", corr,"\n")
```

#### Answer of (c): Draw a scatter plot of VDAX log returns vs DAX log returns. Superimpose your linear fit from part (b) to make a diagram exactly like Figure 12 of Lecture 1.

```r
options(repr.plot.height=5, repr.plot.width=8)
plot(as.numeric(log_DAX), as.numeric(log_VDAX), ylab="VDAX log return", xlab="DAX log return",
    main="Log-log plot of VDAX vs DAX",cex.main=0.9,
    col='black',lwd=1)
abline(reg_DAX,col="red",lwd=3)
```
