#### 1. (4 points)

Let $X_i$ be independent power-law distributed random variables, so that for large $x$,

$$
\Pr(X_i>x) \sim \frac{1}{x^{\zeta_{X_i}}} 
$$

 Prove the following properties of the tail exponents $\zeta_{X_i}$:

 (a) $$
\zeta_{X^\alpha}=\frac{\zeta_{X}}{\alpha}%\label{eq:PLpower}
$$

(b) $$
\zeta_{\alpha\,X}=\zeta_{X}
%\label{eq:PLprod}
$$



While x is big enough, it should be positive, therefore we have
$$
\Pr(X^\alpha > x) = \Pr(X > x^{1/\alpha}).
$$

We also know that
$$
\Pr(X^\alpha > x) \sim \frac{1}{x^{\zeta_{X^\alpha}}},
$$
and
$$
\Pr(X > x^{1/\alpha}) \sim \frac{1}{x^\frac{\zeta_X}{\alpha}}.
$$
So we have
$$
\zeta_{X^\alpha}=\frac{\zeta_{X}}{\alpha}%\label{eq:PLpower}
$$

We assume $\alpha > 0 $.

We know that$$
\Pr(\alpha X > x) = \Pr(X > \frac{x}{\alpha}).
$$

We also know that
$$
\Pr(\alpha X > x) \sim \frac{1}{x^{\zeta_{\alpha X}}},
$$
and
$$
\Pr(X > \frac{x}{\alpha}) \sim \frac{1}{x^{\zeta_X}}.
$$
So we have
$$
\zeta_{\alpha X}=\zeta_{X}.
$$

(a) Assuming that SPX daily returns are Student-t distributed with ν=3 , mean  0  and standard deviation of  1% , what is the chance of a daily move of more than 10% ? Of more than 20% ? 


Here we construct another random variable Y, and Y follows student-t distribution with mean = 0, degree of freedom  = 3.


So, the standard deviation of Y is $\sqrt{3}$.


Using X to represent the variable of SPX daily return.


As a result, X follows the same distribution as $0.01*Y/\sqrt{3}$.


So we have:



$$
\mathbb{P}(X>10\%)=\mathbb{P}(0.01Y/\sqrt{3}>10\%)=\mathbb{P}(Y>10\sqrt{3})
$$
$$
\mathbb{P}(X>20\%)=\mathbb{P}(0.01Y/\sqrt{3}>20\%)=\mathbb{P}(Y>20\sqrt{3})
$$

```r
p10 = pt(-10*sqrt(3),df =3)*2
p20 = pt(-20*sqrt(3),df =3)*2
p10
p20
```

The probability that SPX daily movement is greater than 10% is 
0.00041937423065354


The probability that SPX daily movement is greater than 20% is
5.28929179596659e-05

```r
###How many years should we expect to wait to get moves of over 10%  and 20% respectively?
year_10 <- (1/p10) / 252

year_20 <- (1/p20) / 252

format(paste("The number of years we should expect to wait to get move of over 10%  :  ", round(year_10, 10)))
format(paste("The number of years we should expect to wait to get move of over 20%  :  ", round(year_20, 10)))
```

```r
### (b) Repeat the same computation assuming SPX daily returns are normally distributed. Include in your answer all steps in your computation.
### Recall that the SPX dropped  23%  on October 19, 1987.
p10_2 = pnorm(-0.1,mean=0,sd=0.01)*2
p20_2 = pnorm(-0.2,mean=0,sd=0.01)*2
format(paste("The probability that the movement is over 10% under normality assumption is  :  ", round(p10_2, 30)))
format(paste("The probability that the movement is over 20% under normality assumption is  :  ", round(p20_2, 30)))
```

DAX is a German stock index analogous to the SPX index.  VDAX is the volatility index (VIX) of DAX.

(a) Using the lecture code as a template, use the quantmod package to download time series of DAX and VDAX.

(b) Regress VDAX log returns vs DAX log returns. What is the $R^2$ of this regression? What is the correlation? 

(c) Draw a scatter plot of VDAX log returns vs DAX log returns. Superimpose your linear fit from part (b) to make a diagram exactly like Figure 12 of Lecture 1.

```r
#(a) Using the lecture code as a template, use the quantmod package to download time series of DAX and VDAX.
library(quantmod)
library(boot)
```

```r
#Get DAX and VDAX data from Yahoo!
options("getSymbols.warning4.0"=FALSE)

getSymbols("^GDAXI",from="1927-01-01") #Creates the time series object DAX
getSymbols("^VDAX",from="1927-01-01") # Get the time seris data for VDAX
```

```r
head(GDAXI)
tail(GDAXI)
head(VDAX)
tail(VDAX)
```

```r
#quantmod function automatically aligns data from two series
mm <- specifyModel(Cl(GDAXI)~Cl(VDAX))
daxVdaxData <-modelData(mm) 

Vdax <- daxVdaxData[,"Cl.VDAX"]
dax <- daxVdaxData[,"Cl.GDAXI"]

# compute the log returns for vdax
ret.Vdax <- log(Vdax / lag(Vdax))
# remove the NA values
ret.Vdax <- ret.Vdax[!is.na(ret.Vdax)]
# compute the log returns for dax
ret.dax <- log(dax / lag(dax))
# remove the NA values
ret.dax <- ret.dax[!is.na(ret.dax)]

# Regress VDAX log returns vs DAX log returns
Vdax_dax_reg <- lm(ret.Vdax ~ ret.dax)

summary(Vdax_dax_reg)
cor(ret.Vdax, ret.dax)
```

From the result, the  R2R2  of the regression is 0.5133, and the correlation is -0.7166.


```r
plot(as.numeric(ret.dax), as.numeric(ret.Vdax), xlab="DAX log returns",ylab="VDAX log returns")
abline(Vdax_dax_reg,col="red",lwd=2)
```
