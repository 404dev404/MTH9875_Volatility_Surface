If $B_t$ is Brownian motion, show that

$$
X_t=e^{\int_0^t \sigma_s\,dB_s-\frac12\,\int_0^t\,\sigma_s^2\,ds}
$$

is a martingale. (Hint:  Apply Itô's Lemma to $e^Y$ where $Y=\log X$.)

Deduce that

$$
e^{\frac12\,\alpha^2\,t}=\E\left[e^{\alpha\,\sqrt{t}\,Z}\right]
%\label{eq:expmartingale}
$$

where $Z \sim N(0,1)$.  Is this last equation still true if $\alpha$ is a matrix?  Why?

###Solution:
<font color = "purple">
Let $Y_t = \log(X_t)$, where
$$
dY_t = -\frac{1}{2}\sigma_t^2dt + \sigma_tdB_t
$$
Denote $f(t,y) = e^{Y_t}$apply Ito's lemma to $e^{Y_t}$, we have:
$$
df(t,y) = (0-\frac{1}{2}\sigma_t^2e^Y_t + \frac{\sigma_t^2}{2}e^Y_t)dt + \sigma_te^Y_tdBt
$$
$$
= \sigma_te^Y_tdBt
$$
The drift part of the result is 0, which means $f(t,y) = e^{Y_t} = X_t$ is a martingale.

Deduce that
$$
e^{\frac12\,\alpha^2\,t}=\E\left[e^{\alpha\,\sqrt{t}\,Z}\right]
%\label{eq:expmartingale}
$$
Look at the right-hand-side first:
$$
RHS = \E[e^{\alpha\sqrt{t}Z}]
$$
$$
= \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty e^{\alpha\sqrt{t}-\frac{Z^2}{2}}dZ
$$
$$
= \frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty{e^{-\frac{1}{2}(Z^2 - 2\alpha\sqrt{t}+\alpha^2t)+\frac{1}{2}\alpha^2t}dZ}
$$
$$
= e^{\frac{1}{2}\alpha^2t}\frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty e^{-\frac{1}{2}(Z-\alpha\sqrt{t})}dZ
$$
$$
= e^{\frac{1}{2}\alpha^2t}
$$
$$
= LHS
$$

## Strang splitting

### 2. (4 points) 

If $A$ and $B$ are (in general non-commutative) matrices, show that

$$
e^{(A+B)\,\Delta}=e^{A\,\Delta/2}\,e^{B\,\Delta}\,e^{A\,\Delta/2} + \cO(\Delta^3).
$$

<font color = 'purple'>
LHS =
\begin{eqnarray*}
e ^{(A+B)\Delta} &=& 1 + (A+B)\Delta + \frac12 (A+B)^2 \Delta^2 + \cO(\Delta^3).
\end{eqnarray*}

RHS =

\begin{eqnarray*}
e^{A\,\Delta/2}\,e^{B\,\Delta}\,e^{A\,\Delta/2} &=& \big[1 + \frac{A\Delta}{2} + \frac{A^2 \Delta^2}{8} + \cO(\Delta^3) \big]\cdot \big[1 + B\Delta + \frac12 B^2 \Delta^2 + \cO(\Delta^3) \big] \cdot \big[1 + \frac{A\Delta}{2} + \frac{A^2 \Delta^2}{8} + \cO(\Delta^3) \big]\\
&=& \big[ 1 + \frac12 A\Delta + B\Delta + \frac18 A^2\Delta^2 + \frac12 B^2\Delta^2 + \frac12 AB\Delta^2 + \cO(\Delta^3)\big] \cdot \big[1 + \frac{A\Delta}{2} + \frac{A^2 \Delta^2}{8} + \cO(\Delta^3) \big]\\
&=& 1 + (A+B)\Delta + \big(\frac12A + B\big)\Delta\cdot \frac12A\Delta + \big(\frac18 A^2 + \frac12 B^2 + \frac12 AB + \frac18 A^2 \big)\Delta^2 + \cO(\Delta^3)\\
&=& 1 + (A+B)\Delta + \frac12 (A+B)^2\Delta^2 + \cO(\Delta^3).
\end{eqnarray*}

As a result, LHS = RHS.

## Ninomiya-Victoir

### 3. (16 points)

Consider the following variance  process:

<a name="eq:varProcess"></a>(1)
$$
dv_t = -\lambda\,(v_t-\bar v)\,dt+\xi\,v_t\,dZ_t.
%\label{eq:vProcess}
$$

(a)
Let $f(v,t)=\E[g(v_T)|\cF_t]$ be the fair value at time $t$ of some claim on $v_T$.  Apply Itô's Lemma to $f$ to find the infinitesimal generator $ \cL$ of the diffusion such that 

$$
\frac{df}{dt}+ \cL f=0.
$$

(b)
Find first order differential operators $ V_0$ and $V_1$ such that

$$
 \cL= V_0+\frac12\,V_1^2.
$$

(c)  Solve the ODEs 

$$
 \frac{d }{dt} v(t) = V_0 v(t) \text{ and }  \frac{d }{dt} v(t) = V_1 v(t)
$$

with boundary condition $v(0)=v_0$.

(d)
Deduce the form of the Ninomiya-Victoir Monte Carlo time step for the variance process [(1)](#eq:varProcess) over some small time interval from $t$ to $t+\Delta$.

### Solution:
<font color = "purple">
(a)
Because 
$$
dv_t = -\lambda\,(v_t-\bar v)\,dt+\xi\,v_t\,dZ_t.
$$
and $f(v,t)$ is in the form of expectation of $g(v_t)$ conditional on a filtration,
we have that $f(v,t)$ is a martingale.

By Ito's lemma:
$$
df(v,t) = (\frac{\partial f}{\partial t}-\lambda(v_t-\bar v)\frac{\partial f}{\partial v}+\frac{1}{2}\xi^2v_t^2\frac{\partial^2 f}{\partial v\partial v})dt + \xi\,v_t\frac{\partial f}{\partial v}dZ
$$
Because $f(v,t)$ is a martingale, the drift part of the above equation should equal to 0, we have:
$$
\frac{\partial f}{\partial t}-\lambda(v_t-\bar v)\frac{\partial f}{\partial v}+\frac{1}{2}\xi^2v_t^2\frac{\partial^2 f}{\partial v} = 0
$$
Thus,
$$
\frac{\partial f}{\partial t}=\lambda(v_t-\bar v)\frac{\partial f}{\partial v}-\frac{1}{2}\xi^2v_t^2\frac{\partial^2 f}{\partial v}
$$
The RHS of the above equation is the required infinitesimal generator, that is:
$$
\cL f = -\lambda(v_t-\bar v)\frac{\partial f}{\partial v}+\frac{1}{2}\xi^2v_t^2\frac{\partial^2 f}{\partial v}
$$





<font color = 'purple'>
(b)
We know that:

$$
V_1 = \xi\, v \,\partial_v,
$$

then
\begin{eqnarray*}
\frac12 V_1^2 &＝& \frac12 \xi\, v \, \partial_v \, \big(\xi\, v \, \partial_v \big)\\
&=& \frac12 \xi\,v\, \big(\xi\,\partial_v + \xi\,v\, \partial_{v,v}  \big)\\
&=& \frac12 \xi^2\,v\, \big(\partial_v + v\, \partial_{v,v}  \big)\\
&=& \frac12 \xi^2\,v\,\partial_v  +  \frac12 \xi^2\,v^2\,\partial_{v,v}.
\end{eqnarray*}

As a result,  we have 

$$
V_0 = \big[-\lambda\,(v - \bar{v}) - \frac12 \xi^2\,v\,\big]\partial_v,
$$

 
<font color = 'purple'>
(c)  Solve the ODEs
We know that:
$$
V_0 = \big[-\lambda\,(v - \bar{v}) - \frac12 \xi^2\,v\,\big]\partial_v,
$$
As a result,
\begin{eqnarray*}
\frac{d}{dt}v(t) &=& -\lambda\,(v(t) - \bar{v}) - \frac12 \xi^2\,v(t)\\
&=& -\big(\lambda + \frac12 \xi^2\big)\, v(t) + \lambda \bar{v}\\
&=& -\big(\lambda + \frac12 \xi^2\big)\,\Big[v(t) - \frac{\lambda \bar{v}}{\lambda + \frac12 \xi^2}   \Big]
\end{eqnarray*}

Let 
$$
y_t = v(t) - \frac{\lambda \bar{v}}{\lambda + \frac12 \xi^2}\,\,\,\text{and}\,\,\,\alpha = -(\lambda+\frac12 \xi^2), $$
then we have:
$$
\frac{dy_t}{y_t} = \alpha\, dt.
$$
As a result:
$$
y_t = C\,e^{\,\alpha\, t}
$$
Now bring in the boundary condition: $v(0)=v_0$
$$
v(t) = \frac{\lambda \bar{v}}{\lambda + \frac12\xi^2} + \Big[ v(0)-\frac{\lambda \bar{v}}{\lambda + \frac12\xi^2} \Big]e^{-(\lambda + \frac12\xi^2)t}
$$



For the second ODE:
$$
\begin{eqnarray}
\frac{d }{dt} v(t) &=& V_1 v(t)\\
\frac{d }{dt} v(t) &=& \xi v(t) (\frac{d}{dv}v)\\
\frac{d }{dt} v(t) &=& \xi v(t)
\end{eqnarray}
$$

The solution is 

$$
v(t) = C\,e^{\xi t}
$$

With boundary conditions, we get:

$$
v(t) = v(0) e^{\xi t}
$$

<font color = 'purple'> 
(d)

$$v_{t+\Delta} = v_t e^{\xi \Delta}(\bar v'+(v_t-\bar v')\,e^{-(\lambda+\frac12\xi^2)\Delta})$$

where

$$v'=\frac{\lambda \bar v}{\lambda+\frac12\xi^2}$$

```r
evolveAlfonsi <- function(v,x,dt,W1,W2,L=NULL){

        eldt2 <- exp(-lambda*dt/2);

        #Variance process
        vbarp <- vbar - eta^2/(4*lambda);
        psi <- (1-eldt2)/lambda;
        v1 <- v*eldt2 + lambda*vbarp*psi;
        v2 <- (v1 > 0) * v1; # Take v2 = 0 if v1<0, else v2=v1
        par <- sqrt(v2) + eta/2*sqrt(dt)*W2;
        vf <- eldt2*par*par + lambda*vbarp*psi;

        # Log-stock process (Andersen equation (33))
        vvf <- (v+vf > 0) * (v+vf);
        dw <- vvf/2*dt;
        x <- x  - dw/2 + rho2m1*sqrt(dw)*W1 +
            rho/eta*(lambda*dw + vf-v -lambda*vbar*dt) ;
        # Impose martingale constraint
        x <- x - log(mean(exp(x)));
        v <- vf;
        return(cbind(x,v));
}
```

```r
### Solution (a) Amend the code for HestonMC2:
HestonMC2 <- function(params){
    
    is.even <- function(j){as.logical((j+1) %% 2)}
    
    res <- function(S0, T, AK, N, m, evolve,exactVols=NULL)
    {
  
    lambda <<- params$lambda;
    rho <<- params$rho;
    eta <<- params$eta;
    vbar <<- params$vbar;
    v0 <<- params$v;
  
    n <- m*2; #n is number of timesteps = 2*m so we can use Romberg extrapolation
    sqrt2 <- sqrt(2);
    rho2m1 <<- sqrt(1-rho*rho);
    vbarp <<- vbar - eta^2/(4*lambda);
    
    negCount <- 0;
    
    # We use a vertical array, one element per M.C. path
    x <- rep(0,N); v <- rep(1,N)*v0;
    xm <- x; vm <- v;
    W1m <- rep(0,N); W2m <- rep(0,N); 
  
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
        # Two sets of correlated normal random vars.
        
        W1 <- rnorm(N);
        W2 <- rnorm(N);
        W1 <- W1 - mean(W1); W1 <- W1/sd(W1);
        W2 <- W2 - mean(W2); W2 <- W2/sd(W2);
        # Now W1 and W2 are forced to have mean=0 and sd=1
        
        W2p <- W2 - cor(W1,W2)*W1; # Eliminate actual correlation
        W2p <- W2p - mean(W2p); W2 <- W2p/sd(W2p); 
        # Now W1 and W2 have mean=0, sd=1 and correlation=0
        
        L <- rbinom(N, size=1, prob=1/2); # Bernoulli rv for NV step
    
        # Add code for subgrid
        W1m <- W1m + W1/sqrt2; W2m <- W2m + W2/sqrt2; # N(0,1) rv's for subgrid
       
        if (is.even(i)) {
            #print(c(i,mean(W1m),mean(W2m),sd(W1m),sd(W2m),cor(W1m,W2m)));
            resm <- evolve(vm,xm,T/m,W1m,W2m,L);
            xm <- resm[,1];
            vm <- resm[,2];
            W1m <- rep(0,N); W2m <- rep(0,N);
        }
    
        res <- evolve(v,x,T/n,W1,W2,L);
        x <- res[,1];
        v <- res[,2];
        negCount <- negCount +mean(v<0)/n; #Probability of negative variance per path per timestep
    
  }

    S <- S0*exp(x);
    Sm <- S0*exp(xm);

    # Now we have three vectors of final stock prices
  
    M <- length(AK);
    AV <- numeric(M); AVdev <- numeric(M);
    BSV <- numeric(M); BSVH <- numeric(M); BSVL <- numeric(M);
    iv2SD <- numeric(M); bias <- numeric(M);
    AVm <- numeric(M); AVmdev <- numeric(M);
    BSVm <- numeric(M); BSVHm <- numeric(M); BSVLm <- numeric(M);
    iv2SDm <- numeric(M);
    AV1 <- numeric(M); AV1dev <- numeric(M);
    BSV1 <- numeric(M); BSVH1 <- numeric(M); BSVL1 <- numeric(M);
    iv2SDrom <- numeric(M);biasRom <- numeric(M);
  
  # Evaluate mean call value for each path
  for (i in 1:M)
  {
    # 2*m timesteps
    K <- AK[i];
    V <- (S>K)*(S - K); # Boundary condition for European call
    AV[i] <- mean(V);
    AVdev[i] <- sqrt(var(V)/length(V));  
    
    BSV[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i]);
    BSVL[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] - AVdev[i]);
    BSVH[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] + AVdev[i]);
    iv2SD[i] <- (BSVH[i]-BSVL[i]);
    
    # m timesteps
    Vm <- (Sm>K)*(Sm - K); # Boundary condition for European call
    AVm[i] <- mean(Vm);
    AVmdev[i] <- sd(Vm) / sqrt(N);
    BSVm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i]);
    BSVLm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] - AVmdev[i]);
    BSVHm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] + AVmdev[i]);
    iv2SDm[i] <- (BSVH[i]-BSVL[i]);
    
    # Richardson extrapolation estimates 
    V1 <- 2*V - Vm;
    AV1[i] <- mean(V1);
    AV1dev[i] <- sd(V1) / sqrt(N);
    BSV1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i]);
    BSVL1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] - AV1dev[i]);
    BSVH1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] + AV1dev[i]);
    iv2SDrom[i] <- (BSVH1[i]-BSVL1[i]);
    
    if(!is.null(exactVols)) {bias <- BSV-exactVols};
    if(!is.null(exactVols)) {biasRom <- BSV1-exactVols};
  }
  
  
  l.AK <- length(AK)      
  data.out <- data.frame(AK,rep(N,l.AK),rep(2*m,l.AK),BSV,bias,iv2SD,BSVm,BSV1,biasRom,iv2SDrom) 
  names(data.out) <- c("Strikes","Paths","Steps","ivol","bias","twoSd","ivolm", "ivolRichardson", "biasRichardson","twoSdRichardson") 
  out = list("output"=data.out,"log_stock" = x, "vol_int" = v)
  return(out)   
}
return(res)
}
```

```r
### Code for 4(a)
HestonQMC <- function(params){
    
    is.even <- function(j){as.logical((j+1) %% 2)}
    
    res <- function(S0, T, AK, N, m, evolve,exactVols=NULL)
    {
  
    lambda <<- params$lambda
    rho <<- params$rho
    eta <<- params$eta
    vbar <<- params$vbar
    v0 <<- params$v
  
    n <- m*2 #n is number of timesteps = 2*m so we can use Romberg extrapolation
    sqrt2 <- sqrt(2)
    rho2m1 <<- sqrt(1-rho*rho)
    vbarp <<- vbar - eta^2/(4*lambda)
    
    negCount <- 0
    
    # We use a vertical array, one element per M.C. path
    x <- rep(0,N); v <- rep(1,N)*v0
    xm <- x; vm <- v
    W1m <- rep(0,N); W2m <- rep(0,N)
    
    # Generation of quasi random numbers now takes place outside the timestep loop
    Z <- sobol(n=N,dim=n,scram=3,seed=4711,norm=T)
      Zperp <- sobol(n=N,dim=n,scram=3,seed=17,norm=T)
      
    # Loop for bias computation (N small, n big)
    for (i in 1:n)
    {
        # Two sets of correlated normal random vars.
        
        W1 <- Z[,i] #Take ith column of pre-generated quasi-rvs
        W2 <- Zperp[,i]
        W1 <- W1 - mean(W1); W1 <- W1/sd(W1)
        W2 <- W2 - mean(W2); W2 <- W2/sd(W2)
        # Now W1 and W2 are forced to have mean=0 and sd=1
        
        W2p <- W2 - cor(W1,W2)*W1 # Eliminate actual correlation
        W2p <- W2p - mean(W2p); W2 <- W2p/sd(W2p) 
        # Now W1 and W2 have mean=0, sd=1 and correlation=0
        
        L <- rbinom(N, size=1, prob=1/2) # Bernoulli rv for NV step
    
        # Add code for subgrid
        W1m <- W1m + W1/sqrt2; W2m <- W2m + W2/sqrt2 # N(0,1) rv's for subgrid
       
        if (is.even(i)) {
            resm <- evolve(vm,xm,T/m,W1m,W2m,L)
            xm <- resm[,1]
            vm <- resm[,2]
            W1m <- rep(0,N); W2m <- rep(0,N);
        }
    
        res <- evolve(v,x,T/n,W1,W2,L)
        x <- res[,1]
        v <- res[,2]
        
  }

    S <- S0*exp(x)
    Sm <- S0*exp(xm)

    # Now we have three vectors of final stock prices
    M <- length(AK);
    AV <- numeric(M); AVdev <- numeric(M);
    BSV <- numeric(M); BSVH <- numeric(M); BSVL <- numeric(M);
    iv2SD <- numeric(M); bias <- numeric(M);
    AVm <- numeric(M); AVmdev <- numeric(M);
    BSVm <- numeric(M); BSVHm <- numeric(M); BSVLm <- numeric(M);
    iv2SDm <- numeric(M);
    AV1 <- numeric(M); AV1dev <- numeric(M);
    BSV1 <- numeric(M); BSVH1 <- numeric(M); BSVL1 <- numeric(M);
    iv2SDrom <- numeric(M);biasRom <- numeric(M);
  
  # Evaluate mean call value for each path
  for (i in 1:M)
  {
    # 2*m timesteps
    K <- AK[i];
    V <- (S>K)*(S - K); # Boundary condition for European call
    AV[i] <- mean(V);
    AVdev[i] <- sqrt(var(V)/length(V));
    BSV[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i]);
    BSVL[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] - AVdev[i]);
    BSVH[i] <- BSImpliedVolCall(S0, K, T, 0, AV[i] + AVdev[i]);
    iv2SD[i] <- (BSVH[i]-BSVL[i]);
    
    # m timesteps
    Vm <- (Sm>K)*(Sm - K); # Boundary condition for European call
    AVm[i] <- mean(Vm);
    AVmdev[i] <- sd(Vm) / sqrt(N);
    BSVm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i]);
    BSVLm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] - AVmdev[i]);
    BSVHm[i] <- BSImpliedVolCall(S0, K, T, 0, AVm[i] + AVmdev[i]);
    iv2SDm[i] <- (BSVH[i]-BSVL[i]);
    
    # Richardson extrapolation estimates 
    V1 <- 2*V - Vm
    AV1[i] <- mean(V1)
    AV1dev[i] <- sd(V1) / sqrt(N)
    BSV1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i])
    BSVL1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] - AV1dev[i])
    BSVH1[i] <- BSImpliedVolCall(S0, K, T, 0, AV1[i] + AV1dev[i])
    iv2SDrom[i] <- (BSVH1[i]-BSVL1[i])
    
    if(!is.null(exactVols)) {bias <- BSV-exactVols}
    if(!is.null(exactVols)) {biasRom <- BSV1-exactVols}
  }
  
  data.out <- data.frame(AK,rep(N,M),rep(2*m,M),BSV,bias,iv2SD,BSVm,BSV1,biasRom,iv2SDrom)
  names(data.out) <- c("Strikes","Paths","Steps","ivol","bias","2sd","ivolm", "ivolRichardson", "biasRichardson",
                       "2sdRichardson")
  out<-list("output"=data.out,"log_stock"=x,"vol_int"=v)
  return(out) 
  
}
return(res)
}
```

```r
###(b)
evolveAlfonsiF <- function(v,x,dt,W1,W2,L=NULL){

        eldt2 <- exp(-lambda*dt/2);

        #Variance process
        vbarp <- vbar - eta^2/(4*lambda);
        psi <- (1-eldt2)/lambda;
        v1 <- v*eldt2 + lambda*vbarp*psi;
        v2 <- (v1 > 0) * v1; # Take v2 = 0 if v1<0, else v2=v1
        par <- sqrt(v2) + eta/2*sqrt(dt)*W2;
        vf <- eldt2*par*par + lambda*vbarp*psi;

        # Log-stock process (Andersen equation (33))
        vvf <- (v+vf > 0) * (v+vf);
        dw <- vvf/2*dt;
        x <- x  - dw/2 + rho2m1*sqrt(dw)*W1 +
            rho/eta*(lambda*dw + vf-v -lambda*vbar*dt) ;
        # Impose martingale constraint
        x <- x - log(mean(exp(x)));
        v <- vf;
        return(cbind(x,v));
}
```

```r
library(randtoolbox)
paramsBCC <- list(lambda = 1.15,rho = -0.64,eta = 0.39,vbar = 0.04,v = 0.04)
HMC2 <- HestonMC2(paramsBCC)(S0=1, T=1, AK=1.2, N=100000, m=8, evolve=evolveAlfonsiF) 
QMC <- HestonQMC(paramsBCC)(S0=1, T=1, AK=1.2, N=100000, m=8, evolve=evolveAlfonsiF)
```

```r
###(C)
path = seq(1000,100000,1000)
#Define HMC and MC2 function as of N - Number of paths
IV_MC2 = function(N){HestonMC2(paramsBCC)(S0=1, T=1, AK=1.2, N, m=8, evolve=evolveAlfonsiF)$output['ivolRichardson'][1,]}
IV_QMC = function(N){HestonQMC(paramsBCC)(S0=1, T=1, AK=1.2, N, m=8, evolve=evolveAlfonsiF)$output["ivolRichardson"][1,]}
#vectorize the functions in paths
MC_vol <- sapply(path, IV_MC2)
QMC_vol <- sapply(path, IV_QMC)
vol_exact<-impvolHestonRaw(paramsBCC)(k=log(1.2),t=1)
tol_lower = rep(vol_exact - 0.001, length(path))
tol_upper = rep(vol_exact + 0.001, length(path))
#Generate plots
plot(path, type = "l",MC_vol,col='blue',xlab='Paths',ylab='vols %',ylim=c(0.14,0.16))
lines(path, QMC_vol,col='red',ylim=c(0.14,0.16))
#Add lower and upper bounds
lines(path,tol_upper,col='orange')
lines(path,tol_lower,col = 'orange')
legend('topleft',c('MC','QMC'),lty=1,col=c('blue','red'))
```

```r
###(a)
ks = seq(0.6,1.5,0.1)
IV_BS = function(k){impvolHestonRaw(paramsBCC)(log(k),1)}
MCvol = function(k){{HestonMC2(paramsBCC)(S0=1, T=1, AK=k, N=100000, m=8, evolve=evolveAlfonsiF)$output['ivolRichardson'][1,]}}
QMCvol =function(k){{HestonQMC(paramsBCC)(S0=1, T=1, AK=k, N=100000, m=8, evolve=evolveAlfonsiF)$output['ivolRichardson'][1,]}}

IV_BS_arr = sapply(ks,IV_BS)
IV_MC_arr = sapply(ks,MCvol)
IV_QMC_arr = sapply(ks,QMCvol)
plot(ks,IV_MC_arr, col = 'red',lty=2,xlab="strike k",ylab='implied Vol.')
lines(ks,IV_QMC_arr, col = 'green',lty=2)
lines(ks,IV_BS_arr,col = 'blue',lty=2)
legend('topright',c('IV_MC','IV_QMC','IV_BS'),lty=2,col=c('red','green','blue'))
```

```r
x_MC = HMC2$log_stock
v_MC = HMC2$vol_int
x_QMC = QMC$log_stock
v_QMC = QMC$vol_int

local_vol = function(x,x_arr,v_arr){
    max_x = max(x_arr)
    min_x = min(x_arr)
    bin = 150
    tol = (max_x - min_x)/bin
    i = floor((x - min_x)/tol)
    low = min_x + i*tol
    high = min_x + (i+1)*tol
    #print (i)
    #print (low)
    #print(high)
    #Use bins
    indicator = (x_arr > low) * (x_arr < high)
    N = sum(indicator)
    #print(N)
    res = sum(v_arr*indicator)/N
    return (res)
}

xs = seq(-0.5,0.5,0.02)

lv_MC = sapply(xs,function(x){local_vol(x,x_MC,v_MC)})
lv_QMC = sapply(xs,function(x){local_vol(x,x_QMC,v_QMC)})
plot(xs,lv_MC,col='green',type='l',xlab="log k",ylab="vol.",main = "Local Volatility Smile - MC vs QMC")
lines(xs,lv_QMC,col = 'red')
legend('topright',c('Local Vol - MC','Local Vol - QMC'),col= c('green','red'),lty=1)
```

```r
###(c)
Heston_IV_1 = (sapply(exp(xs),MCvol))
Heston_IV_2 = (sapply(exp(xs),QMCvol))
mean_Heston_IV = (Heston_IV_1 + Heston_IV_2)/2
plot(xs,mean_Heston_IV^2, col = 'red',type = 'l',lty=1,xlab="log-strike",ylab='implied Vol.',ylim=c(0.01,0.13))
lines(xs,lv_MC,col = 'green',type = 'l',lty=1)
legend('topright',c('Heston Variance','Local Variance'),lty=1,col = c('red','green'))
```
